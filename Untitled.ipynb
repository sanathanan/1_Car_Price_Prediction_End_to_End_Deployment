```python
# I am creating my final dataset. I removed car_Name which is not needed for car price prediction
final_df = df[['Year', 'Selling_Price', 'Present_Price', 'Kms_Driven', 'Fuel_Type', 'Seller_Type', 'Transmission', 'Owner']]
final_df.head()
```

```python
# I am creating the current year function to calculate the year of the car
final_df['Current_year']=2021
final_df.head()
```

```python
# Finding the total no of years of the car
final_df['No_of_Years']= final_df['Current_year']-final_df['Year']
final_df.head()
```

```python
# I don't need Year and current year now. Because, i already calclated the no of years of the car
final_df.drop(columns=['Year', 'Current_year'], inplace=True)
final_df.head()
```

```python
# I am doing One hot Encoding. (It is basically a dummy variable trap)
final_df = pd.get_dummies(final_df, drop_first=True)
final_df.head()
```

# we are using Hyper parameter tuning with the help of RandomsearchCV to findout the best parameters

```python
import numpy as np

# Number of trees in random forest
n_estimators = [int(x) for x in np.linspace(start = 100, stop = 1200, num = 12)]
# Number of features to consider at every split
max_features = ['auto', 'sqrt']
# Maximum number of levels in tree
max_depth = [int(x) for x in np.linspace(5, 30, num = 6)]
# max_depth.append(None)
# Minimum number of samples required to split a node
min_samples_split = [2, 5, 10, 15, 100]
# Minimum number of samples required at each leaf node
min_samples_leaf = [1, 2, 5, 10]
```

```python
# Create the random grid
random_grid = {'n_estimators': n_estimators,
               'max_features': max_features,
               'max_depth': max_depth,
               'min_samples_split': min_samples_split,
               'min_samples_leaf': min_samples_leaf}

print(random_grid)
```
